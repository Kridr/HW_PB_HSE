```python
handle = Entrez.efetch(db="nucleotide", rettype="gb", retmode="text",
                       id="CP063829.1")
for seq_record in SeqIO.parse(handle, "gb"):
    print (seq_record.id, seq_record.description[:50] + "...")
    print ("Sequence length %i," % len(seq_record))
    print ("%i features," % len(seq_record.features))
    print ("from: %s" % seq_record.annotations["source"])
```

Разделим последовательности на 10 частей

```python
str_seq = str(seq_record.seq)

parts = 10
chunks, chunk_size = len(str_seq), len(str_seq) // parts
str_seq_list = [str_seq[i : i+chunk_size] for i in range(0, chunks, chunk_size)]
```

Посчитаем частоты как для оригинальных последовательностей, так и для перемешанных

```python
from collections import defaultdict
import random

counter = defaultdict(int)
counter_shuf = defaultdict(int)

all_pos_count = sum([len(str_seq_list[i]) - 3 for i in range(parts)])

for i in range(parts):
    for word in ['AAAA', 'AATT', 'ATAT', 'ATTA']:
        counter[word] += str_seq_list[i].count(word)

        l = list(str_seq_list[i])
        random.shuffle(l)
        str_seq_list_i_shuffled = ''.join(l)
        counter_shuf[word] += str_seq_list_i_shuffled.count(word)

freq = {(k, v / all_pos_count) for k, v in counter.items()}
freq_shuf = {(k, v / all_pos_count) for k, v in counter_shuf.items()}
```
